```python
import tensorflow as tf
import cv2
import os
import matplotlib.pyplot as plt
import numpy as np
import PIL.Image
import os
import h5py  
import torch
```

```python
model_yawn = torch.load('yawn_3.h5')
model_eye = tf.keras.models.load_model('eye_model.h5')
img_size = 224
eye_prediction=[]
```

```python
def prepare(filepath, face_cas="haarcascade_frontalface_default.xml"):
    img_array = cv2.imread(filepath, cv2.IMREAD_GRAYSCALE)
    backtorgb = cv2.cvtColor(img_array,cv2.COLOR_GRAY2RGB)
    new_array = cv2.resize(backtorgb, (img_size,img_size))
    X_input = np.array(new_array).reshape(1, img_size, img_size, 3)
    X_input = X_input/255.0
    return X_input
```

```python
IMG_SIZE = 145
def prepare1(filepath, face_cas="haarcascade_frontalface_default.xml"):
    image_array = cv2.imread(filepath, cv2.IMREAD_COLOR)
    resized_array = cv2.resize(image_array, (IMG_SIZE, IMG_SIZE))
    
    X = np.array(resized_array)
    X = (X.reshape(X.shape[0] * X.shape[1] * X.shape[2]))
    return X
```

```python
def Open_Close(prediction):
    for i in range(len(prediction)):
        if prediction[i]<0:
            return 0
            
    return 1
```

```python
def comb(address):
    if read_eyes(address)==0:
        return True#sleeply
    return False
```

```python
def get_path(path):
    list_p=[]
    list_set=[]
    counter=0
    for img in os.listdir(path):
        list_p.append(img)
        counter+=1
        if counter==5:
            list_set.append(list_p)
            list_p=[]
            counter=0
    return list_set
```

```python
import cv2
import matplotlib.pyplot as plt
```

```python
# 载入人脸识别和眼睛识别的两个xml文件
face_xml = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
eye_xml = cv2.CascadeClassifier('haarcascade_eye.xml')
```

```python
def read_eyes(address):
    test_IMG = address
    test_img = plt.imread(test_IMG)
    eye_prediction=[]
    # 灰度处理
    test_gray = cv2.cvtColor(test_img, cv2.COLOR_BGR2GRAY)
    # 人脸识别
    face = face_xml.detectMultiScale(test_gray, 1.3, 2)  # 参数：1、灰度图片， 2、缩放比例， 3、阈值
    # 绘制出识别到的人脸
    for (x, y, w, h) in face:
    #     cv2.waitkey(0)
        # 在人脸的基础上识别眼睛
        face_gray = test_gray[y:y+h, x:x+w]
        face_color = test_img[y:y+h, x:x+w]
        # 眼睛识别
        eyes = eye_xml.detectMultiScale(face_gray)
        #将识别出的图片进行保存
        im = PIL.Image.open(test_IMG)  # 用PIL打开一个图片
        for i in range(len(eyes)):
            if eyes[i][2]>70:
                if eyes[i][0]>220 or eyes[i][0]<80:
                    box = (face[0][0]+eyes[i][0], face[0][1]+eyes[i][1], face[0][0]+eyes[i][0]+eyes[i][2], face[0][1]+eyes[i][1]+eyes[i][3])  # box代表需要剪切图片的位置格式为:xmin ymin xmax ymax
                    ng = im.crop(box)  # 对im进行裁剪 保存为ng(这里im保持不变)
                    ng.save('eye_test/copy'+str(i)+'.jpg')
                    eye_p = model_eye.predict([prepare('eye_test/copy'+str(i)+'.jpg')])[0][0]
                    eye_prediction.append(eye_p)
    return Open_Close(eye_prediction)
```
