```python
Datadirectory = "train"
Classes = ["Closed","Open"]
for category in Classes:
    path = os.path.join(Datadirectory, category)
    for img in os.listdir(path):
        img_array = cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE)
        backtorgb = cv2.cvtColor(img_array,cv2.COLOR_GRAY2RGB)
        plt.imshow(img_array, cmap="gray")
        plt.show()
        break
    break
    
```

```python
img_size = 224
new_array= cv2.resize(backtorgb, (img_size,img_size))
plt.imshow(new_array, cmap="gray")
plt.show()
```

```python
img_array.shape
```

```python
training_Data = []

def create_training_Data():
    for category in Classes:
        path = os.path.join(Datadirectory, category)
        class_num = Classes.index(category)
        for img in os.listdir(path):
            try:
                img_array = cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE)
                backtorgb = cv2.cvtColor(img_array,cv2.COLOR_GRAY2RGB)
                new_array = cv2.resize(backtorgb, (img_size,img_size))
                training_Data.append([new_array,class_num])
            except Exception as e:
                pass
```

```python
create_training_Data()
```

```python
print(len(training_Data))
```

```python
import random
random.shuffle(training_Data)
```

```python
X=[]
y=[]

for features,label in training_Data:
    X.append(features)
    y.append(label)
    
X = np.array(X).reshape(-1,img_size, img_size,3)
    
```

```python
X=X/255.0;
```

```python
import keras
from tensorflow.keras import layers
from tensorflow.keras import backend as K
```

```python
def dice_coef(y_true, y_pred, smooth=1):
    y_true_f = K.flatten(y_true)
    y_pred_f = K.flatten(y_pred)
    intersection = K.sum(y_true_f * y_pred_f)
    return (2. * intersection + smooth) / (K.sum(y_true_f) + K.sum(y_pred_f) + smooth)
```

```python
model = tf.keras.applications.mobilenet.MobileNet()
```

```python
model.summary()
```

```python
base_input = model.layers[0].input
base_output = model.layers[-4].output
Flat_layer = layers.Flatten()(base_output)
final_output = layers.Dense(1)(Flat_layer)
final_ouput = layers.Activation('sigmoid')(final_output)
```

```python
new_model = keras.Model(inputs = base_input, outputs = final_output)
new_model.summary()
```

```python
new_model.compile(loss="binary_crossentropy", optimizer = "adam", metrics = ["accuracy"])
```

```python
history=new_model.fit(X,Y, epochs = 5, validation_split = 0.1)
```

```python
new_model.save('eye_model.h5')
```

```python
def prepare(filepath, face_cas="haarcascade_frontalface_default.xml"):
    img_array = cv2.imread(filepath, cv2.IMREAD_GRAYSCALE)
    backtorgb = cv2.cvtColor(img_array,cv2.COLOR_GRAY2RGB)
    new_array = cv2.resize(backtorgb, (img_size,img_size))
    X_input = np.array(new_array).reshape(1, img_size, img_size, 3)
    X_input = X_input/255.0
    return X_input
```

```python
import cv2
import matplotlib.pyplot as plt
```

```python
# 载入人脸识别和眼睛识别的两个xml文件
face_xml = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
eye_xml = cv2.CascadeClassifier('haarcascade_eye.xml')
```

```python
test_IMG = 'DATASET/4578.jpg'
test_img = plt.imread(test_IMG)
plt.imshow(test_img)
plt.show()
```

```python
# 灰度处理
test_gray = cv2.cvtColor(test_img, cv2.COLOR_BGR2GRAY)
```

```python
# 人脸识别
face = face_xml.detectMultiScale(test_gray, 1.3, 2)  # 参数：1、灰度图片， 2、缩放比例， 3、阈值
print("这张图片中有%d张人脸" % len(face))
```

```python
import PIL.Image
import os
```

```python
# 绘制出识别到的人脸
for (x, y, w, h) in face:
    print(face)
    cv2.rectangle(test_img, (x, y), (x+w, y+h), (0, 255, 0), 2)  # 绘制人脸方框
    plt.imshow(test_img)
#     cv2.waitkey(0)
    # 在人脸的基础上识别眼睛
    face_gray = test_gray[y:y+h, x:x+w]
    face_color = test_img[y:y+h, x:x+w]
    # 眼睛识别
    eyes = eye_xml.detectMultiScale(face_gray)
    print("在这张脸上有%d个眼睛" % len(eyes))
    print(eyes)
    # 绘制出识别到的眼睛
    for (e_x, e_y, e_w, e_h) in eyes:
        cv2.rectangle(face_color, (e_x, e_y), (e_x+e_w, e_y+e_h), (0, 255, 0), 2)  # 绘制眼睛方框
```

```python
# 绘制出识别到的人脸
for (x, y, w, h) in face:
    cv2.rectangle(test_img, (x, y), (x+w, y+h), (0, 255, 0), 2)  # 绘制人脸方框
    plt.imshow(test_img)
#     cv2.waitkey(0)
    # 在人脸的基础上识别眼睛
    face_gray = test_gray[y:y+h, x:x+w]
    face_color = test_img[y:y+h, x:x+w]
    # 眼睛识别
    eyes = eye_xml.detectMultiScale(face_gray)
    print("在这张脸上有%d个眼睛" % len(eyes))
    #将识别出的图片进行保存
    im = PIL.Image.open(test_IMG)  # 用PIL打开一个图片
    for i in range(len(eyes)):
        if eyes[i][2]>70:
            if eyes[i][0]>220 or eyes[i][0]<80:
                box = (face[0][0]+eyes[i][0], face[0][1]+eyes[i][1], face[0][0]+eyes[i][0]+eyes[i][2], face[0][1]+eyes[i][1]+eyes[i][3])  # box代表需要剪切图片的位置格式为:xmin ymin xmax ymax
                ng = im.crop(box)  # 对im进行裁剪 保存为ng(这里im保持不变)
                ng.save('eye_test/copy'+str(i)+'.jpg')
                eye_p = new_model.predict([prepare('eye_test/copy'+str(i)+'.jpg')])[0][0]
                eye_prediction.append(eye_p)
    # 绘制出识别到的眼睛
    for (e_x, e_y, e_w, e_h) in eyes:
        cv2.rectangle(face_color, (e_x, e_y), (e_x+e_w, e_y+e_h), (0, 255, 0), 2)  # 绘制眼睛方框
```
