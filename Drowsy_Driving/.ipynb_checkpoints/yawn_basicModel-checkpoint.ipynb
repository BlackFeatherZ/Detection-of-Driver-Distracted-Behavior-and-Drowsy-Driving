```python
def face_for_yawn(direc="train", face_cas_path="haarcascade_frontalface_default.xml"):
    yaw_no = []
    IMG_SIZE = 145
    categories = ["yawn", "no_yawn"]
    for category in categories:
        path_link = os.path.join(direc, category)
        class_num1 = categories.index(category)
        print(class_num1)
        for image in os.listdir(path_link):
            image_array = cv2.imread(os.path.join(path_link, image), cv2.IMREAD_COLOR)
            resized_array = cv2.resize(image_array, (IMG_SIZE, IMG_SIZE))
            yaw_no.append([resized_array, class_num1])
    return yaw_no
```

```python
yawn_no_yawn = face_for_yawn()
```

```python
X = []
y = []
for feature, label in yawn_no_yawn:
    X.append(feature)
    y.append(label)
```

```python
X[0]
```

```python
X = np.array(X)
X = (X.reshape(X.shape[0], X.shape[1] * X.shape[2] * X.shape[3]))
```

```python
X
```

```python
from sklearn.preprocessing import LabelBinarizer
label_bin = LabelBinarizer()
y = label_bin.fit_transform(y)
```

```python
y = np.array(y)
```

```python
from sklearn.model_selection import train_test_split
seed = 42
test_size = 0.30
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=seed, test_size=test_size)
```

```python
len(X_test)
```

```python
#Importing librairies

import pandas as pd 
import numpy as np

# Scikit-learn library: For SVM
from sklearn import preprocessing
from sklearn.metrics import confusion_matrix
from sklearn import svm
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier

import itertools

# Matplotlib library to plot the charts
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab

# Library for the statistic data vizualisation
import seaborn

%matplotlib inline
```

```python
#use LogisticRegression model
lr = LogisticRegression()
lr.fit(X_train,y_train)
predlr = lr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(y_test, predlr))
print('\n')
print(classification_report(y_test, predlr))
lr.score(X_test,y_test)
```

```python
#use other model:
clf = svm.SVC(kernel='linear')
clf.fit(X_train,y_train)
predsvm=clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(y_test, predsvm))
print('\n')
print(classification_report(y_test, predsvm))
clf.score(X_test,y_test)
```

```python
# 梯度提升决策树GradientBoostingClassifier
# 训练、测试数据集
from sklearn.ensemble import GradientBoostingClassifier
model= GradientBoostingClassifier(n_estimators=100, learning_rate=1.0, max_depth=1, random_state=0)
model.fit(X_train, y_train)
predicted= model.predict(X_test)
```

```python
# 作出混淆矩阵、分类报告以及成功率
# 分类报告：macro avg：宏平均值（不考虑权重） weighted avg：微平均值（考虑权重）
# 一般来说，微平均值更具有参考意义
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(y_test, predicted))
print('\n')
print(classification_report(y_test, predicted))
model.score(X_test,y_test)
```

```python
#use KNN model
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X_train, y_train) 
predsknn=neigh.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(y_test, predsknn))
print('\n')
print(classification_report(y_test, predsknn))
neigh.score(X_test,y_test)
```
